```python
df_games = pd.read_csv("data/df_games.csv")
df_teams = pd.read_csv("data/df_teams.csv")

df_games=df_games[(df_games['game_id'].notna())]

df_games['game_id']=df_games['game_id'].astype(int).astype(str)

#df_games.sort_values(['season', 'season'], inplace=True)
#df_games = df_games.groupby(['season', 'season']).first().reset_index(drop=True)

df_games.head()
```

```python
def getGameData(df_games):
    """
    Takes list of game-ids and extracts all data.
    Return dataframe with event-information and one dataframe with a summary of all games
    """
    
    data=[]
    datasummary=[]
    
    for index, row in df_games.iterrows():
            
        url = 'http://stats.swehockey.se/Game/Events/' + row['game_id']
        
        df_gamedata = pd.read_html(url)
        
        # Get home and away data
        home = df_gamedata[2].columns[0].split("\xa0-\xa0")[0]
        away = df_gamedata[2].columns[0].split("\xa0-\xa0")[1]
        
        # Get relevant columns, rename and only keep rows with time (length: 5)
        try:
            if np.asarray(df_gamedata[5].columns.values)[[0][0]][0]=='Actions':
                df_gameevents = df_gamedata[5].iloc[:,[0,1,2,3,4]] 
        except:
            pass
                
        try:
            if np.asarray(df_gamedata[10].columns.values)[[0][0]][0]=='Actions':
                df_gameevents = df_gamedata[10].iloc[:,[0,1,2,3,4]] 
        except:
            pass
           
        df_gameevents.columns = ['time', 'event', 'team', 'players', 'on_ice']
        
        df_gameevents = df_gameevents[df_gameevents['time'].str.len()==5]
        df_gameevents['game_id'] = row['game_id']
        df_gameevents['home'] = home
        df_gameevents['away'] = away
        
        # Get gamessummary
        df_gamesummary = df_gamedata[2].iloc[:,[1,2,5,6]]
        df_gamesummary = df_gamesummary.iloc[[1]]
        df_gamesummary.columns = ['shots_h', 'shots_h_period', 'shots_a', 'shots_a_period']
        df_gamesummary['league'] = df_gamedata[2].iloc[0,3]
        df_gamesummary['game_id'] = row['game_id']
        df_gamesummary['home'] = home
        df_gamesummary['away'] = away
        
        data.append(df_gameevents)
        datasummary.append(df_gamesummary)
        
        #print('collected ', index, game_id)
            
    games=pd.concat(data)
    gamessummary = pd.concat(datasummary)

    
    return games, gamessummary

swehockey_team_mapping = {
'MODO': 'MODO Hockey', 
'SAIK':'Skellefteå AIK',
'AIK':'AIK',
'ASP':'Asplöven HC',
'TIK':'Timrå IK',
'BIF':'Brynäs IF',
'VÄX':'Växjö Lakers HC',
'PAN':'IK Pantern',
'RBK':'Rögle BK',
'FBK':'Färjestad BK',
'HV71':'HV 71',
'FRÖ':'Frölunda HC',
'LHC':'Linköping HC',
'LHF':'Luleå HF',
'LIF':'Leksands IF',
'ÖHK':'Örebro HK',
'BIK':'BIK Karlskoga',
'BOR':'Borås HC',
'VIK':'Västerås IK',
'VIT':'HC Vita Hästen',
'VÄS':'Västerviks IK',
'IKO':'IK Oskarshamn',
'SUN':'IF Sundsvall Hockey',
'BJÖ':'IF Björklöven',
'MIK':'Mora IK',
'MIF':'IF Malmö Redhawks',
'AIS':'Almtuna IS',
'TRO':'IF Troja/Ljungby',
'DIF':'Djurgårdens IF',
'KAR':'Karlskrona HK',
'TIN':'Tingsryds AIF',
'SKE':'Skellefteå AIK',
'SSK':'Södertälje SK',
'FHC':'Frölunda HC',
'Mora':'Mora IK',
'Troja':'IF Troja/Ljungby',
'IFMRe':'IF Malmö Redhawks',
'BoHC':'Borås HC',
'NVIF':'Väsby IK HK',
'VLHC':'Växjö Lakers HC',
'BoIK':'BIK Karlskoga',
'IFBj':'IF Björklöven',
'HIK':'Huddinge IK',
'IFSu':'IF Sundsvall Hockey',
'Mbois':'Mariestad BoIS HC',
'NyHK':'Nyköpings HK',
'LÖV':'IF Björklöven',
'ALM':'Almtuna IS',
'SUND':'IF Sundsvall Hockey',
'HUD':'Huddinge IK',
'HIF':'Hammarby IF',
'BOF':'BIK Karlskoga',
'NYB':'Nybro Vikings IF'
}
```

```python
df_events, df_summary = getGameData(df_games)
```

```python
#df_events['team'].replace(swehockey_team_mapping, inplace=True)
df_events['team_long'] = df_events['team'].map(swehockey_team_mapping)

df_events.head()
```

```python
df_events['event_type'] =np.nan

df_events.loc[df_events['event'].str.contains("min",na=False), 'event_type'] = 'penalty'
df_events.loc[df_events['on_ice'].str.contains("Pos. Part",na=False, case=False), 'event_type'] = 'goal'

# Penalty Info
df_events['penalty_type']=df_events['on_ice'].str.replace('[^A-Za-z]', '')
df_events.loc[df_events['event_type']!='penalty','penalty_type'] =np.nan

# Score info
df_events['score_type']=df_events['event'].str.extract('.*\((.*)\).*')
df_events.loc[df_events['event_type']!='goal','score_type'] =np.nan
df_events['score']=df_events['event'].str.split(" ").str[0]
df_events.loc[df_events['event_type']!='goal','score'] =np.nan

# Forward fill with current score
df_events.sort_values(['game_id', 'time'], inplace=True)
df_events.loc[:,['score']] = df_events.loc[:,['score']].ffill()
df_events.loc[:,['score']] = df_events.loc[:,['score']].fillna('0-0')
df_events[['score_home','score_away']]=df_events['score'].str.split("-", expand=True)

df_events['team_game_status'] ='equal'
df_events.loc[(df_events['team_long']==df_events['team_long']) & (df_events['score_home']>df_events['score_away']),'team_game_status'] ='leading'
df_events.loc[(df_events['team_long']==df_events['team_long']) & (df_events['score_away']>df_events['score_home']),'team_game_status'] ='leading'
df_events.loc[(df_events['team_long']==df_events['team_long']) & (df_events['score_home']<df_events['score_away']),'team_game_status'] ='trailing'
df_events.loc[(df_events['team_long']==df_events['team_long']) & (df_events['score_away']<df_events['score_home']),'team_game_status'] ='trailing'

#df_events.head(n=30)
```

```python

df_penalty_group = df_events[df_events['event_type']=='penalty'].groupby(['game_id', 'team_long', 'penalty_type']).count()

df_penalty_group.reset_index(inplace=True)
df_penalty_group=df_penalty_group[['game_id', 'team_long', 'penalty_type', 'event']]

df_penalty_group.head(n=5)

#df_penalty_group['team'].unique()
```

```python
df_penalty_wide = df_penalty_group.pivot(index=['game_id', 'team_long'],columns='penalty_type', values='event').fillna(0).reset_index()

df_penalty_wide = df_penalty_wide.rename(columns={'team_long': 'team'})


df_penalty_wide.head()
```

```python
# Join onto team data
df_teams['game_id']=df_teams.loc[df_teams['game_id'].notna(),'game_id'].astype(int).astype(str)
df_teams_penalty = pd.merge(df_teams, df_penalty_wide, on=['game_id', 'team'], how='left')


#df_teams_penalty[df_teams_penalty['game_id']=='490388'].head()
```
